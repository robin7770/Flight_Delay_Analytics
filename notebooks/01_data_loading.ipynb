```python
df = pd.read_csv("reportscleaned_flights.csv")
```

```python
print("\nChecking Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
# Top Airlines by Number of Flights
import matplotlib.pyplot as plt
import seaborn as sns
airline_summary = df.groupby('AIRLINE').agg({
    'ARR_DELAY': 'mean',
    'FL_DATE': 'count'
}).rename(columns={'FL_DATE': 'Total_Flights'})

# Filter airlines with enough flights
airline_summary = airline_summary[airline_summary['Total_Flights'] > 10000]

# Sort by delay
airline_summary = airline_summary.sort_values('ARR_DELAY', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(
    x=airline_summary.head(10).index,
    y=airline_summary.head(10)['ARR_DELAY']
)

plt.title("Top 10 Airlines with Highest Average Arrival Delay",
          fontsize=14, fontweight='bold')
plt.ylabel("Average Arrival Delay (Minutes)")
plt.xlabel("Airline")
plt.xticks(rotation=45)
plt.show()
```

```python
# Flight Distribution by Month
# This visualization shows seasonal patterns. 
# Delay increases during peak travel months and adverse weather periods.
monthly_delay = df.groupby('Month')['ARR_DELAY'].mean().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_delay.index, y=monthly_delay.values, marker='o')

plt.title("Seasonal Trend in Average Arrival Delays",
          fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Average Arrival Delay (Minutes)")
plt.grid(alpha=0.3)
plt.show()
```

```python
hourly_delay = df.groupby('Hour')['ARR_DELAY'].mean()

plt.figure(figsize=(10,5))
sns.lineplot(x=hourly_delay.index, y=hourly_delay.values)

plt.title("Average Arrival Delay by Departure Hour",
          fontsize=14, fontweight='bold')
plt.xlabel("Departure Hour")
plt.ylabel("Average Delay (Minutes)")
plt.grid(alpha=0.3)
plt.show()
```

```python
avg_delay = df.groupby('AIRLINE')['ARR_DELAY'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
avg_delay.head(10).plot(kind='bar')
plt.title("Average Arrival Delay by Airline")
plt.ylabel("Average Delay (Minutes)")
plt.xticks(rotation=45)
plt.show()
```

```python
hourly_delay = df.groupby('Hour')['ARR_DELAY'].mean()

plt.figure(figsize=(8,5))
hourly_delay.plot()
plt.title("Average Delay by Departure Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Average Arrival Delay")
plt.show()
```
